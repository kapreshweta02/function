Q.1 What is the difference between a function and a method in Python?

Ans-
Function

A function is a block of code that is defined using def and can be called independently. It's not tied to any object.

Example:

    def greet(name):
    return f"Hello, {name}!"
    print(greet("Alice"))
    
Output:

    Hello, Alice!

Method

A method is also a function, but it is associated with an object—specifically, an instance of a class. It's defined inside a class and is called on the object (or instance) of that class.

Example:

    class Greeter:
    def greet(self, name):
    return f"Hello, {name}!"
    greeter = Greeter()
    print(greeter.greet("Bob"))  
Output:

    Hello, Bob!

Here, greet is a method because it's tied to the Greeter class and called with an object of that class.

Quick comparison

    Feature	        Function	         Method
    
    Defined using	  def keyword	      def inside a class
    
    Associated with	Not tied to any  	Bound to an object
                       object
    
    Called with	    function_name  	  object.method_name(args)
                       (args)

Q.3 What are the different ways to define and call a function in Python?

Ans-In Python, you can define and call functions in several different ways, depending on your needs.

1.Standard Function Defination

2.Function with default arguments

3.Function with keyword arguments

4.Function with Arbitrary positional arguments

5.Function with Arbitrary keyword arguments

6.Lambda Function

7.Function as an argument


1.Standard Function Defination

A normal function created using the def keyword to perform a task.

Example:

    def greet(name):
    return f"Hello, {name}!"
    greet("Alice")  
    
Output:

    Hello, Alice!
2.Function with Default Arguments

A function where some parameters have default values if no argument is passed for them.

Example:

    def greet(name="Guest"):
    return f"Hello, {name}!"

    greet()         # Output: Hello, Guest
    greet("Alice")  # Output: Hello, Alice

3.Function with Keyword Arguments

Function where arguments are passed by specifying the parameter name

Example:

    def describe_pet(animal, name):
    return f"{name} is a {animal}."
    describe_pet(animal="dog", name="Buddy")

Output:

    Buddy is a dog.
4.Function with Arbitrary Positional Arguments (*args)

Function that accepts any number of positional arguments, packed into a tuple.

Example:

    def add_all(*numbers):
    return sum(numbers)
    add_all(1, 2, 3, 4)
Output:

    10
5.Function with Arbitrary Keyword Arguments (**kwargs)

Function that accepts any number of named arguments, packed into a dictionary.

Example:

    def print_info(**info):
      for key, value in info.items():
        print(f"{key}: {value}")

    print_info(name="Alice", age=30)
Output:

    name: Alice
    age: 30

6.Lambda (Anonymous) Functions

A small, unnamed function defined with the lambda keyword.

Example:

    square = lambda x: x * x
    square(5)  
Output:

    25
7.Function as an Argument (Higher-Order Function)

A function that takes another function as a parameter.

Example:

    def apply_func(func, value):
      return func(value)

    def double(x):
      return x * 2

    apply_func(double, 4)
    
Output:

    8

Q.5 What are iterators in Python and how do they differ from iterables?

An iterator is an object in Python that allows you to traverse through all the elements of a collection (like a list or string) one element at a time.

It remembers where it is during iteration.

It uses two main methods:

1.__iter__() → returns the iterator object itself.

2.__next__() → returns the next item and moves to the next position.

*What are Iterables in Python?

An iterable is any Python object that can be looped over (iterated over) — like a list, tuple, string, or dictionary.

An iterable doesn’t do the actual iteration by itself.

It must use iter() to create an iterator first.

Example to Understand Both

Iterable Example:

    my_list = [1, 2, 3]
     for item in my_list:  # my_list is iterable
    print(item)

    my_list = [1, 2, 3]

    print(next(iterator))  # 1
    print(next(iterator))  # 2
    print(next(iterator))  # 3

Here, iterator manually fetches elements one by one.

Q.6 Explain the concept of generators in Python and how they are defined.

Ans-
Generator

A generator is a special type of iterator that generates values on the fly, one at a time, using the yield keyword.

Instead of returning all the values at once like a list, it pauses after each value and resumes where it left off.

To Define a Generator

Using a Function with yield

    def count_up_to(n):
       i = 1
       while i <= n:
       yield i
       i += 1
Using a Generator Expression (like a lazy list comprehension)

    gen = (x * x for x in range(5))
How Generators Work:

When you call a generator function, it doesn’t run immediately — it returns a generator object.

You can:

*Use next() to get values one at a time.

*Use it in a loop (like for) to iterate automatically.

Example:

    def numbers():
       yield 1
       yield 2
       yield 3
    gen = numbers()

    print(next(gen))  # 1
    print(next(gen))  # 2
    print(next(gen))  # 3
**next(gen) now will raise StopIteration**

Why Use Generators?

    Benefit	                     Why It’s Useful
    
    Memory efficient	     Doesn’t store all values in memory
    
    Lazy evaluation	      Computes values only when needed

    Infinite sequences	   Can generate endless series (e.g., Fibonacci)


Q.7 What are the advantages of using generators over regular functions?

Ans-

**Advantages of Generators over Regular Functions**
1.Memory Efficiency

Generators don’t store the entire result in memory — they yield one item at a time.

Example:

Instead of creating a list of 1 million numbers, a generator will produce them one by one as needed.

    def count_up_to(n):
       i = 1
       while i <= n:
          yield i
          i += 1
**Great for large datasets**

2.Lazy Evaluation

Generators only compute values when needed, which saves time and resources.

Use case:

You may not need all items — maybe just the first 10 from a million.

3.Better Performance in Loops

Less overhead compared to creating full data structures like lists.

Example:

    for number in (x * x for x in range(1000000)):
      if number > 100:
        break
**The loop stops early, and the generator doesn’t do unnecessary work.**

4.Infinite Sequences

Generators can produce infinite values without running out of memory.

Example:

    def infinite_numbers():
       num = 1
       while True:
          yield num
          num += 1
5.Cleaner Code for Iterators

Generators simplify the process of writing iterators using yield instead of creating a full class with __iter__() and __next__().

6.Pipeline Compatibility

Generators can be chained together to build data pipelines — each stage processes items one at a time.



Q.8 What is a lambda function in Python and when is it typically used?

Ans-Lambda function

A lambda function is a small, anonymous (nameless) function defined using the lambda keyword.

It can take any number of arguments, but has only one expression.

It returns the result of that expression automatically (no need for return).

Syntax:

    lambda arguments: expression
Example:

    square = lambda x: x * x
    print(square(5))   
Output:
    25
This does the same thing as:

    def square(x):
      return x * x
When is a Lambda Function Typically Used

Lambda functions are commonly used for short, throwaway functions, especially when you:

1.Use it temporarily

    print((lambda x: x + 2)(3))  
Output:

    5
2.Use it as an argument to other functions

With map():

    nums = [1, 2, 3]
    squared = list(map(lambda x: x ** 2, nums))
Output:

    [1, 4, 9]

With filter():

    even = list(filter(lambda x: x % 2 == 0, nums))
Output:

    [2]
With sorted():

    points = [(1, 2), (3, 1), (5, 0)]
    sorted_points = sorted(points, key=lambda x: x[1])
**sorts by the second value**

3.Use it in functional programming or GUI callbacks

Example:

    button.onClick(lambda: print("Clicked!"))
**Limitations of Lambda:**

1.Only one expression (no loops, if/else blocks).

2.Not suitable for complex logic.

3.Not as readable as regular functions for larger tasks.



Q.9 Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is used to apply a function to each item of an iterable (like a list or tuple) and returns a map object, which is an iterator containing the results.

Syntax:

    map(function, iterable)
*function ->The function to apply to each item.

*iterable ->A list, tuple, or any iterable data structure.

    def double(x):

      [2, 4, 6, 8]

    numbers = [1, 2, 3, 4]
    result = map(lambda x: x * x, numbers)

    print(list(result))

    [1, 4, 9, 16]

    a = [1, 2, 3]

    result = map(lambda x, y: x + y, a, b)

     [5, 7, 9]

->Transforms data in a list or other iterable.

->Simplifies code compared to writing loops.

->Works great with lambda functions for quick tasks.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

my_list = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(my_list)
print("Sum of even numbers:", result)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
fib = FibonacciIterator(10)

for num in fib:
    print(num, end=' ')

```

6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def generate_powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for value in generate_powers_of_two(5):
    print(value, end=' ')
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
import os

def read_file_line_by_line(filename):
    absolute_path = os.path.abspath(filename)
    print(f"Absolute path of file: {absolute_path}")
    if not os.path.exists(filename):
        print(f"Error: File '{filename}' not found.")
        return

    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line(r"C:\Users\ASUS\Desktop\example.txt"):
    print(line)
```

9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

# Formula: (Celsius × 9/5) + 32 = Fahrenheit
fahrenheit = list(map(lambda temp: (temp * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
```

10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))

# Example usage
input_string = "Hello World"
result = remove_vowels(input_string)
print("String without vowels:", result)
```
